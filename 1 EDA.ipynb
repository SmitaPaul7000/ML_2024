**Feature Scaling(NORMALIZATION & STANDARDIZATION)**

```python
hr_df = pd.read_csv('Human_Resources.csv')
hr_df.head(2)
```

```python
hr_df['Age'].values
```

```python
hr_df['Age'].values.reshape(-1,1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
hr_df['Age'] = scaler.fit_transform(hr_df['Age'].values.reshape(-1,1))
```

```python
hr_df['Age'].values
```

```python
hr_df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler_standard = StandardScaler()
hr_df['Age'] = scaler_standard.fit_transform(hr_df['Age'].values.reshape(-1,1))
print("Standardized Data:\n", hr_df['Age'])
```
